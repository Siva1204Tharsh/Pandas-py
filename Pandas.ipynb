### 05.Data slicing 

```python
## 01.title
print(df_csv.columns) # to get the column names
print(df_csv['Attack']) # to get the data based on specific column name
print(df_csv[['Attack','Defense']]) # to get the data based on specific column names
print(df_csv['Name'][0:6]) ## to get data bsed on specific coloumn name and specific row index
```

```python
### 02.iloc -integer location
print(df_csv.iloc[0]) # first row
print(df_csv.iloc[:,0]) # df_csv.iloc[startRow:endRow,startColumn:endColumn]
print(df_csv.iloc[0:2,0:2]) # first two rows and first two columns
```

```python
## 03.Iterrows
for index, row in df_csv.iterrows():
    print(index, row["Name"])
```

```python
## 04.Filter -Loc
print(df_csv.loc[df_csv['Attack']>=50])
```

### 09.Fix NaN values

```python
## 01. acquiring Dataset without NaN by removing that row (If String)
df_csv=df_csv.dropna(subset=['Type 2']) ## Mehods  -1
df_csv=df_csv[df_csv['Type 2'].notnull()] ## Mehods  -2
df_csv.isnull().any()
```

```python
## 02.Filling NaN values with the Mean/Median of that specific Row (If Numerical Values)
meanDataset=df_csv.fillna(df_csv['HP'].mean()) ## Methods1
medianDataset=df_csv.fillna(df_csv['HP'].median()) ## Methods 2
```
